```python
x_stop = 50
x_step = 1
rectangle_interval = 10
example_size = 10
```

```python
import numpy as np

x_linear = np.arange(0, x_stop, x_step)

print(type(x_linear), x_linear.shape)
print(x_linear[:example_size])
```

```python
random_coefficient = np.random.rand(x_stop//x_step) - 0.5
random_coefficient = random_coefficient * 20

print(type(random_coefficient), random_coefficient.shape)
print(random_coefficient[:example_size])
```

```python
sum = x_linear + random_coefficient
sum = np.sort(sum)

print(sum[:example_size])
```

```python
import matplotlib.pyplot as plt

plt.stem(sum, np.ones(len(sum)))
plt.ylim(0, 2)
plt.show()
```

```python
firing_rate = []

for i in range(0, x_stop, x_step):
    count = 0
    for i1 in sum:
        if i-rectangle_interval <= i1 < i:
            count += 1
    firing_rate.append(count)

firing_rate = np.array(firing_rate)

print(type(firing_rate), firing_rate.shape)
print(firing_rate[:example_size])
```

```python
plt.stem(x_linear, firing_rate)
plt.show()
```
